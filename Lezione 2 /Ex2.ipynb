## <span style="color:red"> 02.1 ##
The first exercise consists in calculating the following integral via Monte Carlo:
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$
first using a uniform distribution in $[0,1]$ then with the importance sampling method.

For the importance sampling the distribution is derived with the method of the cumulative inversion from a function calculated by Taylor expanding (around x = 1) the integrand : $ p(x) = 1-\sqrt{(1-x)}$  ( with x random in $[0,1]$), 
the integrand function for this second method is then $f(x) = \frac{\frac{\pi}{4} \cos\left(\frac{\pi x}{2}\right)}{1 - x} $.

I show a picture of such estimation using a number of throws $M = 10^5$ and a number of blocks $N = 10^2$, errors are calculated with the data blocking method.

```python
import matplotlib.pyplot as plt
import numpy as np

media = np.loadtxt("02.1/data.dat", usecols=0, delimiter=',', unpack='true')
errore = np.loadtxt("02.1/data.dat", usecols=1, delimiter=',', unpack='true')
media1 = np.loadtxt("02.1/data1.dat", usecols=0, delimiter=',', unpack='true')
errore1 = np.loadtxt("02.1/data1.dat", usecols=1, delimiter=',', unpack='true')

x = np.arange(100)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].errorbar(x, media, yerr=errore,color='b', ecolor='r', elinewidth=1, capsize=2)
axs[0].set_xlabel('#blocks')
axs[0].set_ylabel('I')
axs[0].set_title('I estimate with uniform distributuion')
axs[0].grid(True)

axs[1].errorbar(x, media1, yerr=errore1, color='b', ecolor='orange', elinewidth=1, capsize=2)
axs[1].set_xlabel('#blocks')
axs[1].set_ylabel('I')
axs[1].set_title('I estimate with importance sampling')
axs[1].grid(True)

plt.tight_layout()

plt.show()
```

As shown above both the methods are able to reproduce the integral value within the statistical uncertainty, note that the one with uniform distribution is reached in $2\sigma$. The correct implementation of the data blocking method is confirmed by the behavior of the error that follows the data fluctuations.


## <span style="color:red"> 02.1 ##
This exercise is about reproducing the 3D random walk on a cubic lattice and in the continuum. Both the routes always start at the origin and:
1. for the cubic lattice ($a = 1$) the walker a make step at the discrete time in the forward or backward direction of length a in one of the 3 directions of the lattice (x,y,z). The step is chosen by extracting a random number that selects if the step should be x-forward, x-backward and so on for the three possibilities
2. for the continuum the walker, at the discrete time, make a step of length $a = 1$ along a random direction obtained by sampling uniformly the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$. While $\phi$ is simply drawn from a random in $[0,2\pi]$, $\theta$ needs to account for the Jacobian of the transformation in spherical coordinates $\sin(\theta)d\theta$ that leads to $\theta = \arccos(1-2a)$ with a uniform in $[0,1]$.

I show a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ for both random walks and, separately, the statistical uncertainty behavior obtained with $M = 10^5$ simulations each with $N = 10^2$ steps, as functions of the RW step. 

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

N = 100
x = np.arange(N) + 1
rms, error = np.loadtxt("02.2/data.dat", usecols=(0,1), delimiter=',', unpack='true')
rmscont, errorcont = np.loadtxt("02.2/datacont.dat", usecols=(0,1), delimiter=',', unpack='true')

def fit_function(N, A):
    return A * np.sqrt(N)

params, covariance = curve_fit(fit_function, x, rms)
paramscont, covariancecont = curve_fit(fit_function, x, rmscont)

A_fit = params[0]
Acont_fit = paramscont[0]

rms_fit = fit_function(x, A_fit)
rmscont_fit = fit_function(x, Acont_fit)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].errorbar(x, rms, yerr=error*10, label="Data", color='b', ecolor='r', elinewidth=1, capsize=3)
axs[0, 0].plot(x, rms_fit, label=f"Fit: A = {A_fit:.2f} * sqrt(N)", color='red', linestyle='--')
axs[0, 0].set_title("RMS for RW on cubic lattice")
axs[0, 0].set_xlabel("Steps")
axs[0, 0].set_ylabel("RMS")
axs[0, 0].grid(True)
axs[0, 0].legend()

axs[0, 1].errorbar(x, rmscont, yerr=errorcont*10, label="Data", color='g', ecolor='orange', elinewidth=1, capsize=3)
axs[0, 1].plot(x, rmscont_fit, label=f"Fit: A = {Acont_fit:.2f} * sqrt(N)", color='red', linestyle='--')
axs[0, 1].set_title('RMS for RW on continuum')
axs[0, 1].set_xlabel('Steps')
axs[0, 1].set_ylabel('RMS')
axs[0, 1].grid(True)
axs[0, 1].legend()

axs[1, 0].plot(x, error, label='Error', color='b')
axs[1, 0].set_title('Error for RW on cubic lattice')
axs[1, 0].set_xlabel('Steps')
axs[1, 0].set_ylabel('Error')
axs[1, 0].grid(True)
axs[1, 0].legend()

axs[1, 1].plot(x, errorcont, label='Error', color='g')
axs[1, 1].set_title('Error for RW on continuum')
axs[1, 1].set_xlabel('Steps')
axs[1, 1].set_ylabel('Error')
axs[1, 1].grid(True)
axs[1, 1].legend()

plt.tight_layout()

plt.show()
```

As suggested i fitted the RMS with $f(N)=k\sqrt{N}$ and the fitting parameter obtained is $1.00$, this is symptom of a diffusive behavior and it means the walker is indeed performing a random walk in both scenario. Also as expected the RMS error grows similarly like $\sqrt{N}$, this is because as the step number grows also grows the spread of possible displacements making the RMS estimate noisier.
