```python
import numpy as np
import matplotlib.pyplot as plt

# Load city coordinates
cities = np.loadtxt("cap_prov_ita.dat")
x, y = cities[:, 0], cities[:, 1]

# Load best routes and convert to 0-based indexing
best_route_L1 = np.loadtxt("best_route_L1.dat", dtype=int) - 1
best_route_L2 = np.loadtxt("best_route_L2.dat", dtype=int) - 1

# Validate indices
for route, label in [(best_route_L1, "L1"), (best_route_L2, "L2")]:
    if np.any(route >= len(cities)) or np.any(route < 0):
        raise ValueError(f"best_route_{label} contains invalid indices after conversion.")

# Ensure routes are closed
if best_route_L1[0] != best_route_L1[-1]:
    best_route_L1 = np.append(best_route_L1, best_route_L1[0])
if best_route_L2[0] != best_route_L2[-1]:
    best_route_L2 = np.append(best_route_L2, best_route_L2[0])

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 8))

# Plot L1 route
axs[0].scatter(x, y, c='red', s=20, label='Cities')
axs[0].plot(x[best_route_L1], y[best_route_L1], 'o-', markersize=4, label='Best Route (L1)', color='blue')
for i, (xi, yi) in enumerate(zip(x, y)):
    axs[0].text(xi + 0.2, yi + 0.2, str(i), fontsize=8)
axs[0].set_title("Best TSP Route (L1 Norm)")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].legend()
axs[0].axis('equal')
axs[0].grid(True)

# Plot L2 route
axs[1].scatter(x, y, c='red', s=20, label='Cities')
axs[1].plot(x[best_route_L2], y[best_route_L2], 's--', markersize=4, label='Best Route (L2)', color='green')
for i, (xi, yi) in enumerate(zip(x, y)):
    axs[1].text(xi + 0.2, yi + 0.2, str(i), fontsize=8)
axs[1].set_title("Best TSP Route (L2 Norm)")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].legend()
axs[1].axis('equal')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
